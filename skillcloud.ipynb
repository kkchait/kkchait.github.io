```python
import wordcloud
import json
import itertools
import matplotlib.pyplot as plt
```

```python
with open('_data/mydata.json', 'r') as fp:
    mydata = fp.read()

mydata = json.loads(mydata)
skill_details = {d['header']: d['skills'] for d in mydata['skillset']['details'] if not d['skip']}

print(skill_details)
```

```python
all_skills = list(itertools.chain.from_iterable(skill_details.values()))
skillcloud = wordcloud.WordCloud(width=800,
                                height=400,
                                background_color=None,
                                mode='RGBA').generate(' '.join(all_skills))
plt.imshow(skillcloud)
```

```python
skillcloud.to_file('assets/img/skillcloud.png')
```
